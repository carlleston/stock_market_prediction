In this notebook we will prepare the data, transforming ',' decimal to '.', M to Million(10‚Å∂) and the getting the dummies variables news, for this last one we will use a forward fill, that will keep the values through the dates until the another scenarios appears.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set()
import datetime
```

```python
df = pd.read_csv("Data//ITUB4.csv",sep=",",decimal=",")
df.columns = ['Date','Close','Open','High','Low', 'Vol','Var']
df['Date'] = pd.to_datetime(df['Date'], format = '%d.%m.%Y')
df.shape
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%d.%m.%Y')
```

```python
df.head()
```

```python
### features news
df_news = pd.read_csv("Data//itau_news_text.csv",sep=",",decimal=",")
df_news.columns = ['index','Date','Link','Class','Text']
df_news=df_news[['Date','Class']]
df_news['Date'] = pd.to_datetime(df_news['Date'], format = '%Y-%m-%d')
df_news.head()
```

```python
result = pd.merge(df, df_news, on='Date', how='outer')
result['Class'] = result['Class'] .fillna(method='bfill')
result = result.dropna()
```

## IBOVESPA

```python
df_IBOVESPA = pd.read_csv("Data//features_Ibovespa.csv",sep=",",decimal=",")
df_IBOVESPA=df_IBOVESPA[['Data','Abertura']]
df_IBOVESPA.columns = ['Date', 'Open_IBOVESPA']
df_IBOVESPA['Date'] = pd.to_datetime(df_IBOVESPA['Date'], format = '%d.%m.%Y')
```

```python
result = pd.merge(result, df_IBOVESPA, on='Date', how='outer')
result = result.dropna()
```

```python
result.tail(15)
```

```python
df = result.copy()
```

```python
df.Vol = (df.Vol.replace(r',', '.', regex=True).replace(r'[KM]+$', '', regex=True).astype(float) * \
            df.Vol.str.extract(r'[\d\.]+([KM]+)', expand=False)
            .fillna(1)
            .replace(['K','M'], [10**3, 10**6]).astype(int))
df['Var'] = df["Var"].str.replace('%','',regex=True).replace(',','.',regex=True).astype(float)
df['Open_IBOVESPA'] = df['Open_IBOVESPA'].apply(lambda x: x.replace('.','').replace(',','.'))
```

```python
df['DayofWeek'] = df['Date'].dt.dayofweek
df['Weekofyear'] = df['Date'].dt.week
df['Quarter'] = df['Date'].dt.quarter
df['Month'] = df['Date'].dt.month
```

```python
df = pd.get_dummies(df, columns=["Class"], prefix=["News"])
```
